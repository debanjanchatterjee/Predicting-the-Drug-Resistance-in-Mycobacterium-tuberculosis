```python
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn import metrics
import pandas
```

```python
X = pandas.read_csv('X_trainData_column_modified_AMK.csv', delimiter=',')

Y = pandas.read_csv('final_X_testData_column_modified_AMK.csv', delimiter=',')
```

```python
clf = svm.SVC(kernel='rbf', probability=True)

clf.fit(X[X.columns[0:219]], X[X.columns[219]])
y_pred = clf.predict_proba(Y[Y.columns[0:219]])
y_pred
```

```python
submission_data = pandas.read_csv('Y_testData_1_nolabels_AMK.csv')
ids=submission_data.iloc[:,:1].values
result=list()

for i in range(len(ids)):
    result.append([ ids[i,0], y_pred[i,1]])

result.insert(0, ['ID','AMK'])
```

```python
output = pandas.DataFrame(result)
output.to_csv('AMK_results.csv',index = False)
```
